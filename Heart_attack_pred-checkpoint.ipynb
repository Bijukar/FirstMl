```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('heart.csv')
df
```

```python
#see basic information about dataset

df.info()
```

```python
#to see null values in dataset
df.isnull().sum()
```

```python
#to see duplicate value 
df.duplicated().sum()
```

```python
#perform One-Hot-Encoding for change categorical to numerical 
ohe=OneHotEncoder(sparse=False,drop='first',dtype=np.int32)
col=['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']
encode_data=ohe.fit_transform(df[col])
```

```python
names=ohe.get_feature_names_out()
d=pd.DataFrame(encode_data,columns=names)
df=pd.concat([df,d],axis=1)
df.drop(col,axis=1,inplace=True)

```

```python
x=df.drop('HeartDisease',axis=1)
y=df['HeartDisease']
```

```python
#split data in train and test data

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#perform Standardisation 
std=StandardScaler()
std.fit(x_train)
new_xtrain=std.transform(x_train)
new_xtest=std.transform(x_test)
```

```python
x_train=pd.DataFrame(new_xtrain,columns=x_train.columns)
x_test=pd.DataFrame(new_xtest,columns=x_test.columns)
```

```python
x_train.describe().round(2)
```

```python
#Train the model
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print('Accuracy score of model= ', accuracy)
```
